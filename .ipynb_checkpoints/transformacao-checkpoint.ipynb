```python
import pandas as pd
import pandera as pa
```

```python
valores_ausentes = ['**', '****', '*****', '###!', '####', 'NULL', '']
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=";", parse_dates = ["ocorrencia_dia"], dayfirst = True, na_values = valores_ausentes)
df
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_cidade":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo":pa.Column(pa.String, nullable=True),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes":pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
#ocorrências no dataframe cuja classificação seja INCIDENTE GRAVE ou estado == SP
filtro1 = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 | filtro2]
```

```python
#ocorrências no dataframe cuja classificação seja INCIDENTE GRAVE ou classificação INCIDENTE e estado == SP
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#Outra forma de filtrar com melhor semântica ocorrências no dataframe cuja classificação seja INCIDENTE GRAVE ou classificação INCIDENTE e estado == SP
filtro1 = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE','INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#ocorrências cuja cidade começem com C
filtro = df.ocorrencia_cidade.str[0] == 'C'
df.loc[filtro]
```

```python
#ocorrências cuja cidade terminem com A
filtro = df.ocorrencia_cidade.str[-1] == 'A'
df.loc[filtro]
```

```python
#ocorrências cuja cidade termine com os caracteres MA
filtro = df.ocorrencia_cidade.str[-2:] == 'MA'
df.loc[filtro]
```

```python
#ocorrências cuja cidade contenha (em qualquer parte do conteúdo) os caracteres MA
filtro = df.ocorrencia_cidade.str.contains('MA')
df.loc[filtro]
```

```python
#ocorrências cuja cidade contenha (em qualquer parte do conteúdo) os caracteres MA ou AL
filtro = df.ocorrencia_cidade.str.contains('MA|AL')
df.loc[filtro]
```

```python
#ocorrencias cujo ano seja 2015
filtro = df.ocorrencia_dia.dt.year == 2015
df.loc[filtro]
```

```python
#ocorrencias cujo ano seja 2015 e mês 12 (método dt)
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
df.loc[filtro1 & filtro2]
```

```python
#outra forma de filtrar ocorrencias cujo ano seja 2015 e mês 12 (método dt)
filtro = (df.ocorrencia_dia.dt.year == 2015) & (df.ocorrencia_dia.dt.month == 12)
df.loc[filtro]
```

```python
#ocorrencias cujo ano seja 2015 e mês 12 e dia 8(método dt)
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
filtro3 = df.ocorrencia_dia.dt.day == 8

df.loc[filtro1 & filtro2 & filtro3]
```

```python
#ocorrencias cujo ano seja 2015 e mês 12 entre os dias 3 e dia 8(método dt)
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
filtro_dia_inicio = df.ocorrencia_dia.dt.day > 2
filtro_dia_fim = df.ocorrencia_dia.dt.day < 9
df.loc[filtro1 & filtro2 & filtro_dia_inicio & filtro_dia_fim]
```

```python
#criando uma nova coluna com o dia (convertido pra string para conseguir concatenar) e a hora
#convertendo para data usando pd.to_datetime
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
```

```python
#ocorrencias cujo ano seja 2015 e mês 12 entre os dias 3 e dia 8(método dt)
filtro1 = df.ocorrencia_dia_hora.dt.year == 2015
filtro2 = df.ocorrencia_dia_hora.dt.month == 12
filtro_dia_inicio = df.ocorrencia_dia_hora.dt.day > 2
filtro_dia_fim = df.ocorrencia_dia_hora.dt.day < 9
df.loc[filtro1 & filtro2 & filtro_dia_inicio & filtro_dia_fim]
```

```python
#filtrando só ocorrencias pela nova coluna de dia e hora
filtro1 = df.ocorrencia_dia_hora >= '2015-12-03 11:00:00'
filtro2 = df.ocorrencia_dia_hora <= '2015-12-08 14:30:00'
df.loc[filtro1 & filtro2]
```

```python
#ocorrencias do ano 2015 e mês 3
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 3
df201503 = df.loc[filtro1 & filtro2]
df201503
```

```python
df201503.count()
```

```python
#agrupando por ocorrencia e fazendo a contagem dos dados não nulos
df201503.groupby(['codigo_ocorrencia']).count()
```

```python
#agrupando pela classificação e fazendo a contagem pelo código de ocorrência
df201503.groupby(['ocorrencia_classificacao']).codigo_ocorrencia.count()
```

```python
#agrupando pela classificação e fazendo a contagem pelo aerodromo.GERA CONTAGEM ERRADA, CUIDADO!
df201503.groupby(['ocorrencia_classificacao']).ocorrencia_aerodromo.count()
```

```python
#agrupando pela classificação e fazendo a contagem pelos registros totais (evitando a contagem incorreta)
df201503.groupby(['ocorrencia_classificacao']).size()
```

```python
#agrupando pela classificação e fazendo a contagem na ordem crescente
df201503.groupby(['ocorrencia_classificacao']).size().sort_values()
```

```python
#agrupando pela classificação e fazendo a contagem na ordem decrescente
df201503.groupby(['ocorrencia_classificacao']).size().sort_values(ascending=False)
```

```python
#ocorrencias na região sudeste no ano de 2010 
filtro1 = df.ocorrencia_dia.dt.year == 2010
filtro2 = df.ocorrencia_uf.isin(['SP','MG','RJ','ES'])
dfsudeste2010 = df.loc[filtro1 & filtro2]
dfsudeste2010
```

```python
#agrupando por classificação e obtendo os registros totais de ocorrências na região sudeste em 2010
dfsudeste2010.groupby(['ocorrencia_classificacao']).size()
```

```python
#agrupando por classificação e uf, obtendo os registros totais de ocorrências na região sudeste em 2010
dfsudeste2010.groupby(['ocorrencia_classificacao','ocorrencia_uf']).size()
```

```python
#agrupando por cidade, obtendo os registros totais de ocorrências em ordem decrescente na região sudeste em 2010
dfsudeste2010.groupby(['ocorrencia_cidade']).size().sort_values(ascending=False)
```

```python
#mostra as ocorrências com total de recomendações maior do que zero na cidade RIO DE JANEIRO
filtro1 = dfsudeste2010.ocorrencia_cidade == 'RIO DE JANEIRO'
filtro2 = dfsudeste2010.total_recomendacoes > 0
dfsudeste2010.loc[filtro1 & filtro2]
```

```python
#soma do total de recomendações na cidade RIO DE JANEIRO
filtro = dfsudeste2010.ocorrencia_cidade == 'RIO DE JANEIRO'
dfsudeste2010.loc[filtro].total_recomendacoes.sum()
```

```python
#mostra a soma do total de recomendações agrupado por cidade
dfsudeste2010.groupby(['ocorrencia_cidade']).total_recomendacoes.sum()
```

```python
#mostra a soma do total de recomendações agrupado por cidade para as que possuem recomendação
filtro = dfsudeste2010.total_recomendacoes > 0 
dfsudeste2010.loc[filtro].groupby(['ocorrencia_cidade']).total_recomendacoes.sum()
```

```python
#mostra a soma do total de recomendações agrupado por aerodromo, incluindo os dados NA
dfsudeste2010.groupby(['ocorrencia_aerodromo'], dropna=False).total_recomendacoes.sum()
```

```python
#mostra a soma do total de recomendações agrupado por cidade e mês para as que possuem recomendação
filtro = dfsudeste2010.total_recomendacoes > 0 
dfsudeste2010.loc[filtro].groupby(['ocorrencia_cidade', dfsudeste2010.ocorrencia_dia.dt.month]).total_recomendacoes.sum()
```

```python
#verificando se os valores encontrados para cidade de SÃO PAULO estão corretos
filtro1 = dfsudeste2010.total_recomendacoes > 0 
filtro2 = dfsudeste2010.ocorrencia_cidade == 'SÃO PAULO'
dfsudeste2010.loc[filtro1 & filtro2]
```
